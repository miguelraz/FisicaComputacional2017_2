Disco de radio 1, cuadrado $[-1,1]^2$

```julia
L = 20
x = linspace(-1, 1, L+1)
y = linspace(-1, 1, L+1)

intervalos_x = [Interval(x[i],x[i+1]) for i in 1:length(x)-1]
intervalos_y = [Interval(y[i],y[i+1]) for i in 1:length(y)-1]
```

```julia
intervalos_x[1]
```

```julia
intervalos_y[1]
```

```julia
intervalos_x[1] × intervalos_y[1]
```

```julia
M = [x × y for x in intervalos_x, y in intervalos_y]
```

```julia
M[1,1]
```

```julia
typeof(M[1,1])
```

```julia
plot(M[1,1])
```

```julia
p = plot(xlim=(-1,1), ylim=(-1,1), aspect_ratio=:equal)

for i in 1:length(intervalos_x)
    for j in 1:length(intervalos_y)
        
        if intervalos_x[i]^2 + intervalos_y[j]^2 < 1
            plot!(M[i,j], c=:red, label="")
            
        elseif intervalos_x[i]^2 + intervalos_y[j]^2 > 1
            plot!(M[i,j], c=:blue, label="")
            
        else
            plot!(M[i,j], c=:yellow, label="")
        end
    end
end
plot!(x->sqrt(1-x^2), -1, 1)
p
```
