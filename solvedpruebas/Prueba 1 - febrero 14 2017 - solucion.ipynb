## `break` en un `for`

```julia
for i in 1:10
    println(i)
    
    if i > 5
        # println("soy mayor")
        break
    end
        
    println("Estoy al final del cuerpo del for")

end
    
    
println("Estoy después del end del for")
```

```julia
for i in 1:10
    println(i)
    
    if i > 5
        # println("soy mayor")
        continue
    end
        
    println("Estoy al final del cuerpo del for")

end
    
    
println("Estoy después del end del for")
```
